# Question

Find maximal early departures in minutes for each airline. Sort results from largest to smallest. Output airline names.

# Query

```sql
select
    L_AIRLINE_ID.NAME as Airline_Name,
    max(-al_perf.DepDelay)as Max_Early_Departure
from al_perf
join L_AIRLINE_ID
    on al_perf.dot_id_reporting_airline = L_AIRLINE_ID.ID
where al_perf.DepDelay < 0
group by L_AIRLINE_ID.NAME
order by Max_Early_Departure desc;
```

# Result

15 rows returned

# Visualization

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("../csv/mini_project_<problem2_Mingyang_Han_mh2393>.csv") 

# Sort by Max_Early_Departure descending (highest first)
df = df.sort_values("Max_Early_Departure", ascending=False)

# Set figure size
plt.figure(figsize=(14,6))

# Create vertical bar plot with color gradient
sns.barplot(
    x="Airline_Name",
    y="Max_Early_Departure",
    data=df,
    palette="Blues"
)

# Add value labels on top of bars
for index, value in enumerate(df["Max_Early_Departure"]):
    plt.text(index, value + 0.5, str(value), ha='center', va='bottom', color='black')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Add title and axis labels
plt.title("Maximum Early Departure by Airline", fontsize=16)
plt.xlabel("Airline")
plt.ylabel("Max Early Departure (minutes)")

plt.grid(False)
plt.tight_layout()
plt.show()
````
