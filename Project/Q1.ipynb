# Question

Find maximal departure delay in minutes for each airline. Sort results from smallest to largest maximum delay. Output airline names and values of the delay.

# Query

```sql
select l_airline_id.name, max(al_perf.depdelayminutes) as maxdepdelay
from al_perf
join l_airline_id on al_perf.dot_id_reporting_airline = l_airline_id.id
group by l_airline_id.name
order by maxdepdelay asc;
```

# Result

15 rows returned

# Visualization

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV and sort by MaxDepDelay
df = pd.read_csv("../csv/Q1.csv").sort_values("MaxDepDelay")

# Create horizontal bar plot with color gradient
plt.figure(figsize=(12,8))
sns.barplot(x="MaxDepDelay", y="Name", data=df, palette="Reds")

# Add value labels
for index, value in enumerate(df["MaxDepDelay"]):
    plt.text(value + 10, index, str(value), color='black', va='center')

# Set title and labels
plt.title("Maximum Departure Delay by Airline", fontsize=16)
plt.xlabel("Max Departure Delay (minutes)")
plt.ylabel("Airline")

plt.grid(False)
plt.tight_layout()
plt.show()

````

````python
df_heat = df.set_index("Name")
plt.figure(figsize=(6,8))
sns.heatmap(df_heat[["MaxDepDelay"]], annot=True, fmt="d", cmap="Reds", cbar=True)
plt.title("Maximum Departure Delay by Airline")
plt.show()
````
